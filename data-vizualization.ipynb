```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Cleaned_ships_data.csv', delimiter=',')
```

```python
df
```

```python
# 2-D график зависимости средней максимальной загрузки кораблей от длины корабля
```

```python
newDf = df.groupby('length', as_index = False)\
    .aggregate({'dwt' : 'mean'})\
    .sort_values('length')
```

```python
plt.plot(newDf['length'], newDf['dwt'])

plt.title('Dependence of the average maximum load of ships on the length of the ship')
plt.xlabel('Ship leght')
plt.ylabel('Average maximum load of ships')

plt.show()
```

```python
# Распределение количества кораблей по их типу
```

```python
topShips = df['ship_name'].value_counts().head(4)

otherShipsCount = df['ship_name'].value_counts().sum() - topShips.sum()

topShips['Other'] = otherShipsCount
```

```python
plt.figure(figsize=(10, 6))
plt.pie(topShips.values, labels=topShips.index, autopct='%1.1f%%')
plt.title('Distribution of the number of ships by their type')
plt.show()
```

```python
# Гистограмма распределения количества кораблей по годам постройки 
```

```python
asd = df['built_year'].value_counts()
```

```python
countOfShip = pd.DataFrame({'year' : asd.index, 'count' : asd.values})\
        .sort_values(by = 'year', ascending=True)
```

```python
histogram = countOfShip.plot(kind='bar', x='year', y='count', figsize=(10,5), color='blue')

histogram.set_title('Distribution of Ships by Year Built')
histogram.set_xlabel('Year Built')
histogram.set_ylabel('Count of Ships')
```

```python
# 2-D Гистограмма распределения количества кораблей по годам постройки (built_year)
# и максимальному весу груза (dwt)
```

```python
plt.hist2d(df['built_year'], df['dwt'], bins=(20,20), cmap='YlOrRd')
plt.title('Distribution of Ships by Built Year and DWT')
plt.xlabel('Built Year')
plt.ylabel('DWT')
plt.colorbar()
plt.show()
```

```python
dfDataSciense = pd.read_csv('Data Science Salary 2021 to 2023.csv', delimiter=',')
```

```python
dfDataSciense
```

```python
# соотношение зарплат по категориям опыта работы 
```

```python
salaryExp = dfDataSciense.groupby('experience_level', as_index = False)\
    .aggregate({'salary_in_usd' : 'mean'})
```

```python
plt.bar(salaryExp['experience_level'], salaryExp['salary_in_usd'])

plt.title('Dependence of salary on work experience')
plt.xlabel('Experience level')
plt.ylabel('Salary in usd')
```

```python
# Исходя из графика можно понять, что специалисты с бОльшим опытом получают бОльшую зарплату
# Так люди с опытом работы "EX (Experienced)" имеют самую высокую заработную плату 
# На втором месте по доходу идут специалисты "SE (Senior)", на третьем "MI (Mid-level)", 
# на четверотом "EN (Entry-Level)"
```

```python
# Распределение заработной платы специалистов
```

```python
sns.histplot(dfDataSciense['salary_in_usd'], kde=True, color='#c44e52', bins=20)

plt.title('Distribution of salaries of specialists')
plt.xlabel('Salary in usd')
plt.ylabel('Number of specialists')
```

```python
# - Распределение заработной платы специалистов по анализу данных смещено влево:
#   большинство специалистов получают зарплаты ниже среднего.

# - Заметный пик в распределении говорит о концентрации профессионалов в определенном диапазоне зарплат
```

```python
# Распределение количества специалистов в зависимости от конктретной области анализа данных
```

```python
topProf = dfDataSciense['job_title'].value_counts().head(5)

otherProf = dfDataSciense['job_title'].value_counts().sum() - topProf.sum()

topProf['Other'] = otherProf
```

```python
plt.figure(figsize=(20, 8))

plt.pie(topProf.values, labels=topProf.index, autopct='%1.1f%%')
plt.title('Distribution of the number of specialists in different fields')
plt.show()
```

```python
# Из круговой диаграммы видно, что количество специалистов таких профессий как
# Data Scientist, Data Engineer, Data Analyst, ML Engineer преобладает
```

```python
# зависит ли медиан зп от размера компании
```

```python
salarySize = dfDataSciense.groupby('company_size', as_index = False)\
    .aggregate({'salary_in_usd' : 'median'})\
    .sort_values('company_size')
```

```python
plt.bar(salarySize['company_size'], salarySize['salary_in_usd'])
plt.title('Dependence of median salary on company size')
plt.xlabel('Company size')
plt.ylabel('Salary in usd')
```
